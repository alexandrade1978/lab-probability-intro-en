## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
import math
# Calculating probability
white = 10
red = 12
black = 8
total = white + red + black
# taking a white ball
p_w = white / total
print(f"The probability of taking a white ball: {p_w:.4f} or {p_w*100:.2f}%")
#Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
p_w_b = p_w * (black / (total - 1))
print(f"The probability of taking a white ball on first extraction and a black ball on second extraction without replacing the first one: {p_w_b:.4f} or {p_w_b*100:.2f}%")
# Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
p_w_b_rep = p_w * (black / total)
print(f"The probability of taking a white ball on first extraction and a black ball on second extraction replacing the first one: {p_w_b_rep:.4f} or {p_w_b_rep*100:.2f}%")
# Taking 3 balls and them being all different colours.
p_3_diff = ((white / total) * (red / (total - 1)) * (black / (total - 2))) * 6
print(f"The probability of taking 3 balls and them being all different colours: {p_3_diff:.4f} or {p_3_diff*100:.2f}%")
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
p_cloudy = 0.4
p_cloudy_rain = 0.5
p_non_cloudy_rain = 0.2
p_rain = (p_cloudy * p_cloudy_rain) + ((1 - p_cloudy) * p_non_cloudy_rain)
print(f"The probability of raining Saturday: {p_rain:.4f} or {p_rain*100:.2f}%")
```

## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

```python
# Events given
p_a = 0.3
p_b = 0.6
p_a_intersect_b = 0.1
# Verification of statements true or false
# P non A = 0.7?
p_not_a = 1 - p_a
if p_not_a == 0.7:
    print("P non A = 0.7 is True")
else:
    print("P non A = 0.7 is False, P non A =", p_not_a)
# P A union B = 0.6?
p_a_union_b = round(p_a + p_b - p_a_intersect_b, 2)
if p_a_union_b == 0.6:
    print("P A union B = 0.6 is True")
else:
    print("P A union B = 0.6 is False, P A union B =", p_a_union_b)
# P A intercect non B = 0.2?
p_a_intersect_not_b = round(p_a - p_a_intersect_b, 2)
if p_a_intersect_not_b == 0.2:
    print("P A intersect non B = 0.2 is True")
else:
    print("P A intersect non B = 0.2 is False, P A intersect non B =", p_a_intersect_not_b)
# P A union non A = 1? (It will always be 1 for any event A)
p_a_union_not_a = p_a + (1 - p_a)
if p_a_union_not_a == 1:
    print("P A union non A = 1 is True")
else:
    print("P A union non A = 1 is False, P A union non A =", p_a_union_not_a)
# P A intersect union B and non B = 0.4? (It will always be P A for any event A and B)
p_a_intersect_union_b_not_b = p_a
if p_a_intersect_union_b_not_b == 0.4:
    print("P A intersect union B and non B = 0.4 is True")
else:
    print("P A intersect union B and non B = 0.4 is False, P A intersect union B and non B =", p_a_intersect_union_b_not_b)
    print("The intersection of P(A) with 1 will always be the event itself.")
# P non A intersect non B = 0.3?
p_not_a_intersect_not_b = round(1 - p_a - p_b + p_a_intersect_b, 2)
if p_not_a_intersect_not_b == 0.3:
    print("P non A intersect non B = 0.3 is True")
else:
    print("P non A intersect non B = 0.3 is False, P non A intersect non B =", p_not_a_intersect_not_b)
# The probability of the complement of A intersection B = 0.9?
p_not_a_intersect_b = round(1 - p_a_intersect_b, 2)
if p_not_a_intersect_b == 0.9:
    print("The probability of the complement of A intersection B = 0.9 is True")
else:
    print("The probability of the complement of A intersection B = 0.9 is False, The probability of the complement of A intersection B =", p_not_a_intersect_b)
```

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

```python
# Prob Dice rolling
p_italy = 1/6
p_china = 5/6

# Conditional Probabilities
p_covid_given_italy = 0.20
p_covid_given_china = 0.01

# Total Probability of getting Covid
p_covid = (p_italy * p_covid_given_italy) + (p_china * p_covid_given_china)

# Bayes Theorem: P(China | Covid)
p_flew_china = (p_china * p_covid_given_china) / p_covid

print(f"Probability of having flown to China: {p_flew_china:.2f} or {p_flew_china*100:.2f}%")
```

## BONUS

You roll 2 dice. What is the probability that the sum of the numbers is 7? And 8? And 9? Then simulate 10000 times and draw a countplot.

The results in the Countplot are a visual representation of the Theoretical Probability for two dice.

Combinations: The sum of 7 is the most frequent because it has the highest number of combinations (6 out of 36), followed by 8 (5 combinations) and 9 (4 combinations).

Law of Large Numbers: By simulating 10,000 rolls, the random noise is eliminated. This causes the observed frequency to converge almost perfectly with the mathematical expectation.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


total_outcomes = 6 * 6

# Probability for Sum = 7
# Combinations: (1,6), (2,5), (3,4), (4,3), (5,2), (6,1)
prob_7 = 6 / total_outcomes
print(f"The probability of rolling a sum of 7 is: {prob_7:.4f} or {prob_7*100:.2f}%")

# Probability for Sum = 8
# Combinations: (2,6), (3,5), (4,4), (5,3), (6,2)
prob_8 = 5 / total_outcomes
print(f"The probability of rolling a sum of 8 is: {prob_8:.4f} or {prob_8*100:.2f}%")

# Probability for Sum = 9
# Combinations: (3,6), (4,5), (5,4), (6,3)
prob_9 = 4 / total_outcomes
print(f"The probability of rolling a sum of 9 is: {prob_9:.4f} or {prob_9*100:.2f}%")



# Simulate 10,000 rolls for two dice
dice1 = np.random.randint(1, 7, 10000)
dice2 = np.random.randint(1, 7, 10000)
results = dice1 + dice2

sns.set_theme(style="dark")
colors = sns.color_palette("viridis", n_colors=11)
fig, ax = plt.subplots(figsize=(10, 6))

# Set background colors for the whole window and the plot area
fig.patch.set_facecolor('#212529')
fig.set_facecolor('#212529')
ax.set_facecolor('#212529')
# Draw the countplot
ax = sns.countplot(x=results, hue=results, palette="viridis", legend=False)
for patch in ax.patches:
    patch.set_edgecolor('#212529') # Ou use 'none'
    patch.set_linewidth(0)
ax.tick_params(colors='grey')
ax.xaxis.label.set_color('grey')
ax.yaxis.label.set_color('grey')
ax.title.set_color('grey')
# 4. Refine Text and Colors
ax.set_title("Countplot of 10,000 Dice Rolls", color="grey", fontsize=16, pad=20)
ax.set_xlabel("Sum of Two Dice", color="grey", fontsize=12)
ax.set_ylabel("Total Occurrences", color="grey", fontsize=12)

# Make the numbers on the axes (ticks) white
ax.tick_params(colors='grey')

# Optional: Make the grid lines a bit more subtle
ax.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.5)
plt.tight_layout()
plt.show()
```
